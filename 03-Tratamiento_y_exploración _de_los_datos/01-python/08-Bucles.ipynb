## Bucles con while

Bucle infinito

````python
while True:
    print('Estoy dentro del bucle y no puedo salir')
````

````python
numero_grande = -99999999

# Ingresa el primer valor
number = int(input('Introduce un número o escribe -1 para terminar'))

while number != -1:
    # ¿Es el número más grande que el valor numero_grande?

    if number > numero_grande:
        numero_grande = number
    number = int(input('Introduce un número o escribe -1 para terminar'))

print(f'El numeor más grnde es: {numero_grande}')
````

````python
# Imprimir todos los numeors del 1 a n

numero = int(input('Introduce un número'))
i = 1

while i <=numero:
    print(f'Soy el numero {i}')
    i = i + 1

print('Fin del programa')
````

Realizar un programa que solicite n calificaciiones hasta que el usuario de un -1, debe imprimir cuantas fueron aprobatorias y reprobatorias.

````python
cal = int(input('Introduce la calificación'))
calAprob = 0
calRep = 0
while cal != -1:
    if cal >= 7.0 and cal <= 10.0:
        calAprob = calAprob + 1
    elif cal<7.0 and cal >=0:
        calRep += 1
    
    cal = int(input('Introduce la calificación'))

print(f'El numero de aprobados es {calAprob}\nEl numero de reprobados es {calRep}')
````

## Bucle con for

Otro tipo de bucle disponible en `python` proviene de la observación de que a veces es más importante contar los 'giros o vueltas' de bucle que verificar las condiciones.

```python
for i in range (100):
    # do_something()
    pass
```

````python
for i in range(10):
    print(f'El valor de i es {i}')
````

````python
for i in range(10):
    print(f'El valor {i} es {i + 1}')
````

````python
lista = [2,3,4,5,6,7,10]

for li in lista:
    print(li)
````

````python
lista = [2,3,4,5,6,7,10,[9,8,7,6]]

for li in lista:
    print(li)
````

Nota:

El bucle se ha ejecutado 10 veces (es el argumento de la función range()) El valor de la ultima variable de control es 9 (no 10, ya que comienza desde 0 y no desde 1).

````python
for i in range(2,8):
    print(f'El valor de i es {i}')
````

````python
i = 2
while i < 8:
    print(f'El valor de i es {i}')
    i += 1
````

Las sentencias break y continue

 Break - Sale del bucle inmediatamente, e incondicionalmente termina la operación del bucle, el programa empieza a ejecutar las instrucción más cercana después del cuerpo del bucle.

````python
# break
print('la instrucción break')
for i in range(1,6):
    if i == 3:
        break
    print('Dentro del bucle')
print('Fuera del bucle')
````

Continue - Se comporta como si el programa hubiera llegado repentinamente al final del cuerpo; el siguiente turno se inicia y la expresión de condición se prueba de inmediato.

````python
print('\n La instrucción continue: ')
for i in range(1,6):
    if i == 3:
        continue
    print(f'Dentro del bucle. {i}')
print(f'Fuera del bucle')
````
